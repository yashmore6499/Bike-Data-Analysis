```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('BIKE DETAILS.csv')

# Display the first few rows to understand the data
print(df.head())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

```python
# Plot to detect outliers
sns.boxplot(data=df[['selling_price', 'km_driven', 'ex_showroom_price']])
plt.show()

# Optionally remove or treat outliers
# df = df[(df['selling_price'] < df['selling_price'].quantile(0.99))]  # Example
```

```python
# Create new columns
df['car_age'] = 2024 - df['year']
df['price_diff'] = df['ex_showroom_price'] - df['selling_price']

```

```python
# Summary statistics
print(df.describe())
```

```python
# Correlation matrix
corr_matrix = df.select_dtypes(include=['number']).corr()  # Select only numeric columns
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.histplot(df['selling_price'], kde=True)
plt.title('Selling Price Distribution')
plt.show()
```

```python
sns.scatterplot(x='car_age', y='selling_price', data=df)
plt.title('Selling Price vs. Car Age')
plt.show()
```
